```python
from pomap.pomaps.categorical_pomap import CategoricalPomap
import polars as pl
```

```python
example_df = pl.from_dict(
    {'cat1': ['a', 'a', 'b', 'b'],
     'cat2': ['c', 'd', 'c', 'd'],
     'cat3': ['e', 'e', 'e', 'f'],
     }
)
```

```python
example_df
```

```python
pmap_cat1 = CategoricalPomap(column='cat1', labels=['a', 'b'])
pmap_cat2 = CategoricalPomap(column='cat2', labels=['c', 'd'])
pmap_cat3 = CategoricalPomap(column='cat3', labels=['e', 'f'])
```

```python
pmap_prod_12 = pmap_cat1.product(pmap_cat2)
```

```python
pmap_sum = pmap_cat3.sum(pmap_prod_12)
```

```python
pmap_sum.name
```

```python
pmap_sum.view_labels()
```

```python
pmap_cat1.label_rows_as_train(
    example_df,
    label='a'
)
```

```python
pmap_prod_12.label_rows_as_train(
    example_df,
    label={'cat1': 'a', 'cat2': 'c'}
)
```
